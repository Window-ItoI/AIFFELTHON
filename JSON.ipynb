## JSON에서 필요한 정보만 뽑아내는 함수

```python
import pandas as pd

def get_data(json_data, df):
    width = json_data['images'][0]['width']
    height = json_data['images'][0]['height']
    if height == 0:
        height = 1
    rate = width/height
    name = json_data['categories'][0]['name']
    supercategory = json_data['categories'][0]['supercategory']
    Type1 = json_data['categories'][0]['metainfo']['Type1']
    Type2 = json_data['categories'][0]['metainfo']['Type2']
    file_name = json_data['images'][0]['file_name']
    
    df.loc[len(df)] = [file_name, width, height, rate, supercategory, Type1, Type2]
    
    return file_name, width, height, rate, Type2

def get_data_2(json_data, df):
    width = json_data['images'][0]['width']
    height = json_data['images'][0]['height']
    if height == 0:
        height = 1
    rate = width/height
    name = json_data['categories'][0]['name']
    supercategory = json_data['categories'][0]['supercategory']
    Type1 = json_data['categories'][0]['metainfo']['Type1']
    Type2 = json_data['categories'][0]['metainfo']['Type2']
    # file_name = json_data['images'][0]['file_name']
    
    df.loc[len(df)] = [name, width, height, rate, supercategory, Type1, Type2]
    
    return name, supercategory, Type1, Type2
    

df = pd.DataFrame(columns = ['file_name', 'width', 'height', 'rate', 'supercategory', 'Type1', 'Type2'])
df_2 = pd.DataFrame(columns = ['name', 'width', 'height', 'rate', 'supercategory', 'Type1','Type2'])
```

## 폴더 내 JSON 파일의 내용들을 데이터 프레임으로 만들기

* 상위 폴더를 path로 잡고
* 상위 폴더 내 폴더들이 가지고 있는 json 파일에서 원하는 정보들을 가져와 DataFrame으로 만듭니다.

```python
import json
import os

file_list = []
df_count = pd.DataFrame(columns = ['name','supercategory', 'Type1', 'Type2', 'count'])

# folder_path = '/home/aiffel/Mini_AIFFELTHON/랜드마크 이미지/Training/세종특별자치시/'
# folder_path = '/home/aiffel/Mini_AIFFELTHON/랜드마크 이미지/Training/강원도/'
folder_path = '/home/aiffel/Mini_AIFFELTHON/랜드마크 이미지/Validation/제주특별자치도/'


# sample_path = '/home/aiffel/Mini_AIFFELTHON/랜드마크 이미지/Training/강원도/(구)태백등기소/(구)태백등기소_049_38864024.json'

folder_list = os.listdir(folder_path)
# print(folder_list)
for folder in folder_list:
    file_list = os.listdir(folder_path + folder)
    file_count = 0
    for json_file in file_list:
        json_file_path = folder_path + folder + '/' + json_file
        if json_file_path.endswith('.json'):
            with open(json_file_path, 'r') as file:
                json_data = json.load(file)
                file_count += 1
                get_data(json_data, df)
                name, supercategory, Type1, Type2 = get_data_2(json_data, df_2)
    df_count.loc[len(df_count)] = [name, supercategory, Type1, Type2, file_count]
    # file_count 완료
```
